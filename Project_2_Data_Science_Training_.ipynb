```python
import numpy as np
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/inventory_management_dirty.csv')
```

```python
print(df.describe())
```

```python
print("Missing items check:\n", df.isnull().sum())
```

```python
df['Product_ID'].unique()
```

```python
df['Product_ID'] = df['Product_ID'].replace('???', np.nan)
df['Product_ID'] = df['Product_ID'].fillna(df['Product_ID'].mode()[0])
```

```python
df['Product_ID'] = df['Product_ID'].str.strip().str.lower()
df['Product_ID'] = df['Product_ID'].str.replace('p', 'P', regex=False)
```

```python
df['Product_ID'].head(10)
```

```python
df['Product_Name'].unique()
```

```python
df['Product_Name'] = df['Product_Name'].str.strip().str.lower()
df['Product_Name'] = df['Product_Name'].replace('corrupt###', np.nan)
df['Product_Name'] = df['Product_Name'].str.replace(' ', '', regex=False)
```

```python
df['Product_Name'] = df['Product_Name'].fillna(df['Product_Name'].mode()[0])
```

```python
df['Product_Name'] = df['Product_Name'].str.strip().str.title()
```

```python
df['Product_Name'].head(10)
```

```python
df['Category'].unique()
```

```python
df['Category'] = df['Category'].str.strip().str.lower()

df['Category'] = df['Category'].replace({
    'groccery': 'grocery',
    'furnture': 'furniture',
    'clothng': 'clothing',
    'electroncs': 'electronics',
    'elec': 'electronics',
    'stationary': 'stationery',
    'sportss': 'sports'
})
df['Category'] = df['Category'].str.replace(' ', '', regex=False)

df['Category'] = df['Category'].fillna(df['Category'].mode()[0])

df['Category'] = df['Category'].str.title()
```

```python
df['Category'].head(10)
```

```python
df['Supplier'].unique()
```

```python
df['Supplier'] = df['Supplier'].str.strip().str.title()

df['Supplier'] = df['Supplier'].fillna(df['Supplier'].mode()[0])
```

```python
df['Supplier'].head(10)
```

```python
df['Warehouse'].unique()
```

```python
df['Warehouse'] = df['Warehouse'].str.strip()
df['Warehouse'] = df['Warehouse'].replace({
    'delhi': 'Delhi',
    'DELHI': 'Delhi',
    'New Delhi': 'Delhi',
    'Delhi Warehouse': 'Delhi',

    'mumbai': 'Mumbai',
    'MUMBAI': 'Mumbai',
    'Mumbai Warehouse': 'Mumbai',

    'bangalore': 'Bangalore',
    'Bengaluru': 'Bangalore',

    'kolkata': 'Kolkata',
    'KOLKATA': 'Kolkata',

    'chennai': 'Chennai',
    'CHENNAI': 'Chennai'
})

df['Warehouse'] = df['Warehouse'].fillna(df['Warehouse'].mode()[0])
```

```python
df['Warehouse'].head(10)
```

```python
print(df.dtypes)
```

```python
df['Quantity_Stock'].unique()
```

```python
df['Quantity_Stock'] = pd.to_numeric(df['Quantity_Stock'], errors='coerce')
```

```python
df['Quantity_Stock'].dtypes
```

```python
df.loc[df['Quantity_Stock'] < 0, 'Quantity_Stock'] = np.nan
```

```python
df['Quantity_Stock'] = df['Quantity_Stock'].fillna(df['Quantity_Stock'].mean())
```

```python
df['Quantity_Stock'] = df['Quantity_Stock'].astype(int)
```

```python
df['Unit_Price'].dtypes
```

```python
df['Unit_Price'].unique()
```

```python
df['Unit_Price'] = df['Unit_Price'].astype(str)

df['Unit_Price'] = df['Unit_Price'].str.replace('â‚¹', '', regex=False)
df['Unit_Price'] = df['Unit_Price'].str.replace('$', '', regex=False)
df['Unit_Price'] = df['Unit_Price'].str.replace('₹', '', regex=False)
df['Unit_Price'] = df['Unit_Price'].str.replace(',', '', regex=False)

df['Unit_Price'] = pd.to_numeric(df['Unit_Price'], errors='coerce')
```

```python
# Negative prices treated as invalid values
df.loc[df['Unit_Price'] < 0, 'Unit_Price'] = np.nan
df['Unit_Price'] = df['Unit_Price'].fillna(df['Unit_Price'].mean())
```

```python
df['Purchase_Date'].unique()
```

```python
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'], errors='coerce')
```

```python
df['Purchase_Date'] = df['Purchase_Date'].fillna(df['Purchase_Date'].mode()[0])
```

```python
df['Expiry_Date'].unique()
```

```python
df['Expiry_Date'] = pd.to_datetime(df['Expiry_Date'], errors='coerce')
```

```python
df['Expiry_Date'].isna().sum()
```

```python
df['Expiry_Date'].dtypes
```

```python
df['Expiry_Date'] = df['Expiry_Date'].fillna(df['Expiry_Date'].mode()[0])
```

```python
print("Missing items check:\n", df.isnull().sum())
```

```python
df['Reorder_Level'].unique()
```

```python
df.loc[df['Reorder_Level'] < 0, 'Reorder_Level'] = np.nan
```

```python
df['Reorder_Level'].isna().sum()
```

```python
df['Reorder_Level'] = df['Reorder_Level'].fillna(df['Reorder_Level'].median()).astype(int)
```

```python
df['Reorder_Level'].min()
```

```python
df['Damaged_Items'].unique()
```

```python
df['Damaged_Items'] = df['Damaged_Items'].fillna(df['Damaged_Items'].median())
```

```python
df['Sales_Last_Month'].unique()
```

```python
df['Sales_Last_Month'] = df['Sales_Last_Month'].fillna(df['Sales_Last_Month'].median())
```

```python
df['Status'].unique()
```

```python
df['Status'] = df['Status'].astype(str)
df['Status'] = df['Status'].str.strip().str.lower()
```

```python
df['Status'] = df['Status'].replace({
    'available': 'Available',
    'in stock': 'Available',
    'available ': 'Available',

    'out of stock': 'Out of Stock',
    'out stock': 'Out of Stock',
    'outstock': 'Out of Stock',

    '???': np.nan,
    'nan': np.nan
})
```

```python
df['Status'] = df['Status'].fillna(df['Status'].mode()[0])
```
