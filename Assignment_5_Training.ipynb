### Q4 [loc] Use df.loc to select: (a) row at index 0, (b) rows 2 to 5, (c) rows 0–4 but only columns Name, Score, Grade.

```python
# (a) Row at index 0
print("Row at index 0:")
display(df.loc[0])

# (b) Rows 2 to 5
print("\nRows 2 to 5:")
display(df.loc[2:5])

# (c) Rows 0–4 but only columns Name, Score, Grade
print("\nRows 0-4 for Name, Score, Grade columns:")
display(df.loc[0:4, ['Name', 'Score', 'Grade']])
```

### Q5 [iloc] Use df.iloc to select: (a) first 3 rows, (b) last 2 rows, (c) rows at positions 1, 3, 5 and columns at positions 0, 2, 4.

```python
# (a) First 3 rows
print("First 3 rows:")
display(df.iloc[0:3])

# (b) Last 2 rows
print("\nLast 2 rows:")
display(df.iloc[-2:])

# (c) Rows at positions 1, 3, 5 and columns at positions 0, 2, 4
print("\nRows at positions 1, 3, 5 and columns at positions 0, 2, 4:")
display(df.iloc[[1, 3, 5], [0, 2, 4]])
```

### Q6 [Boolean Filter] Filter students where Score > 80. How many students scored above 80?

```python
students_above_80 = df[df['Score'] > 80]
print("Students with Score > 80:")
display(students_above_80)
print(f"\nNumber of students who scored above 80: {len(students_above_80)}")
```

### Q7 [Boolean Filter] Filter students from Delhi who are in CSE department. Show Name, Score, and Grade only.

```python
delhi_cse_students = df[(df['City'] == 'Delhi') & (df['Department'] == 'CSE')][['Name', 'Score', 'Grade']]
print("Students from Delhi in CSE department (Name, Score, Grade):")
display(delhi_cse_students)
```

### Q8 [Boolean Filter] Find all students who Failed (Passed == False). What is their average Score?

```python
failed_students = df[df['Passed'] == False]
print("Failed students:")
display(failed_students)

if not failed_students.empty:
    average_score_failed = failed_students['Score'].mean()
    print(f"\nAverage Score of failed students: {average_score_failed:.2f}")
else:
    print("\nNo students failed.")
```

### Q9 [String Filter] Filter students whose Name contains the string 'a' (case-insensitive) using `.str.contains()`.

```python
students_with_a_in_name = df[df['Name'].str.contains('a', case=False)]
print("Students whose Name contains 'a' (case-insensitive):")
display(students_with_a_in_name)
```

### Q10 [Add Column] Add a new column 'Score_Category': if Score >= 80 → 'Distinction', Score >= 60 → 'Merit', else 'Pass/Fail'. Use `np.where()` or nested condition.

```python
df['Score_Category'] = np.where(df['Score'] >= 80, 'Distinction',
                                np.where(df['Score'] >= 60, 'Merit', 'Pass/Fail'))
print("DataFrame with new 'Score_Category' column:")
display(df)
```

### BONUS Sort the DataFrame by Score in descending order. Reset the index. Show the top 5 students.

```python
top_5_students = df.sort_values(by='Score', ascending=False).reset_index(drop=True).head(5)
print("Top 5 students by Score (descending):")
display(top_5_students)
```
