### **For a 3×4 random integer array (np.random.randint(1,100,size=(3,4))): compute sum across all elements, row-wise sum (axis=1), and column-wise mean (axis=0).**

### **For a 3×4 random integer array (np.random.randint(1,100,size=(3,4))): compute sum across all elements, row-wise sum (axis=1), and column-wise mean (axis=0).**

```python
a = np.random.randint(1,100,size=(3,4))
a
```

```python
print("Sum of all elements:", np.sum(a))
```

```python
print("Row-wise sum:", np.sum(a, axis=1))
```

```python
print("Column-wise mean:", np.mean(a, axis=0))
```

### **Generate 1000 random numbers from a normal distribution (np.random.normal(mean=50, std=10, size=1000)). Compute mean and std. Are they close to 50 and 10?**

```python
random_numbers = np.random.normal(loc=50, scale=10, size=1000)
```

```python
np.mean(random_numbers)
```

```python
np.std(random_numbers)
```
